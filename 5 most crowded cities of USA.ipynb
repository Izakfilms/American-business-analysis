```python
import pandas as pd
```

```python
data = pd.read_csv('List_of_United_States_cities_by_population_2.csv')
```

```python
data.head()
```

### The goal is to identify the most crowded cities in the USA

1. Clean the data set
2. Sort and find out the top 5 cities

```python
data.drop(['2021\nrank', '2020\ncensus', 'Change', '2020 land area', '2020 land area.1', '2020 population density', '2020 population density.1'], axis=1, inplace=True)
```

```python
data.head()
```

```python
data.rename(columns={'2021\nestimate':'estimated_population'}, inplace=True)
```

```python
data.head()
```

```python
data.info()
```

```python
data.estimated_population = data['estimated_population'].replace(['\,'], '', regex=True)
```

```python
data.head()
```

```python
data.info()
```

```python
data['estimated_population'] = data['estimated_population'].astype('int')
```

```python
data.info()
```

```python
data.head()
```

```python
data.groupby('City').sum()
```

```python
data
```

After cleaning and sorting the data set, we find out that the most crowded cities are New York, California, Illinois, Texas, Arizona
