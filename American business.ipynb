### Choosing the needed columns

```python
right_columns_data = raw_data[['BUSINESS NAME', 'DBA NAME', 'ZIP CODE', 'PRIMARY NAICS DESCRIPTION', 'LOCATION']]
right_columns_data.reset_index()
```

### Cleaning the names of the columns

```python
right_columns_data.rename(columns={'CITY': 'city', 'BUSINESS NAME': 'name', 'DBA NAME': 'dba_name', 'ZIP CODE': 'zip', 'PRIMARY NAICS DESCRIPTION': 'description', 'LOCATION': 'location'}, inplace=True)
```

```python
right_columns_data.reset_index(inplace=True)
```

```python
right_columns_data.head()
```

```python
right_columns_data.rename(columns={'CITY': 'city'}, inplace=True)
```

```python
right_columns_data.head()
```

### Checking and dealing with the missing values

```python
right_columns_data.isnull().sum()
```

```python
right_columns_data = right_columns_data.drop('dba_name', axis=1)
```

```python
right_columns_data.head()
```

```python
right_columns_data.isnull().sum()
```

```python
no_missing_values = right_columns_data.dropna(axis=0)
```

```python
no_missing_values.head()
```

```python
no_missing_values.shape
```

```python
no_missing_values.isnull().sum()
```

### Checking and dealing with duplicates

```python
no_missing_values.duplicated().sum()
```

```python
no_duplicated = no_missing_values.drop_duplicates()
```

```python
no_duplicated.duplicated().sum()
```

```python
no_duplicated.head(300)
```

### Checking and fixing string caps

```python
no_duplicated['city'] = no_duplicated['city'].str.upper()
```

```python
no_duplicated.head(300)
```

```python
no_duplicated['name'] = no_duplicated['name'].str.lower()
```

```python
no_duplicated.head()
```

```python
data = no_duplicated
```

### Slicing top crowded city business' 
While exploring the data set I found out that only New York is listed in the data set. For the other cities, we still need to gather the data.

```python
data.reset_index(inplace=True)
```

```python
data.head()
```

```python
new_york = data.loc[data['city'] == 'NEW YORK'].reset_index(drop=True)
new_york.head()
```

### Exporting to CSV
